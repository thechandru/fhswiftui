Python Enums are a way to create a set of named constants. Here's the basic idea:

```python
from enum import Enum

class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
```

Key behaviors:
- `Color.RED` gives you the enum member (not just `1`)
- `Color.RED.value` gives you `1`
- `Color.RED.name` gives you `"RED"`
- They're comparable: `Color.RED == Color.RED` is `True`
- But `Color.RED == 1` is `False` (unless you use special enum types)

For strings:
```python
class Font(Enum):
    title = "title"
    caption = "caption"
```
- `Font.title` is the enum member
- `Font.title.value` is `"title"`

The main "gotcha": you need `.value` to get the actual string, which can be annoying when passing to functions.
